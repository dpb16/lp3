```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# img = cv2.imread('arrow.png' )
# plt.imshow(img)
# plt.title('Original Image')
# plt.show()

img=cv2.imread('arrow.png')
plt.imshow(img)
plt.title('Its original image')
plt.show()
```

```python
# pointA = (20,20)
# pointB = (400,20)
# cv2.line(img, pointA, pointB, (255, 255, 0), thickness=3, lineType=cv2.LINE_AA)
# plt.imshow(img)
# plt.title("Image Line")
# plt.show()

pointa=(20,30)
pointb=(200,80)
cv2.line(img,pointa,pointb,(255,255,255))
plt.imshow(img)
plt.title('line on image')
plt.show()
```

```python
# start_point =(10, 10)
# end_point =(100,100)
# cv2.rectangle(img, start_point, end_point, (225, 24, 255), thickness= 3, lineType=cv2.LINE_8) 
# plt.imshow(img)
# plt.title("Image Rectangle")
# plt.show()

pointa=(50,50)
pointb=(100,100)
cv2.rectangle(img,pointa,pointb,(255,0,0),lineType=cv2.LINE_8)
plt.imshow(img)
plt.title("rectangle")
plt.show()
```

```python
circle_center = (50,50)
radius =20
cv2.circle(img, circle_center, radius, (255, 225, 255), thickness=3, lineType=cv2.LINE_AA) 
plt.imshow(img)
plt.title("Image Circle")
plt.show()
```

```python

center = (120,120)
axis1 = (50,64)
axis2 = (25,75)
img = cv2.imread('arrow.png')  # Reload the original image

cv2.ellipse(img, center, axis1, 45, 0, 360, (255, 255, 0), thickness=3)
cv2.ellipse(img, center, axis2, 90, 0, 360, (0, 255, 255), thickness=3)
plt.imshow(img)
plt.title("Image Ellipse")
plt.show()
```
